# 1.创建画板

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,3,2,4])
```

```python
x = np.linspace(0,2,100)
x
```

```python
fig,ax = plt.subplots()
ax.plot(x,x, label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('title')
ax.legend()
```
